### Характерные различия между двумя группами попробуем визуализировать при помощи гистограмм

*Построим функцию для поочередного вывода на экран столбчатых гистограмм с распределением того факта, клиент попал в отток или нет, в зависимости от различных исходных признаков*:

```python
gym_churn1 = gym_churn.copy(deep=True)
gym_churn1['churn'] = gym_churn1['churn'].replace(1, "попал в отток")
gym_churn1['churn'] = gym_churn1['churn'].replace(0, "не попал в отток")

def hist(column, parameter, parameter1):

    plt.figure(figsize=(8,8))
    
    sns.barplot(y=column, x='churn', hue='churn', data=gym_churn1, estimator=np.mean)
    plt.legend([],[], frameon=False)
    plt.ylabel(f'{parameter1}',fontsize = 15)
    plt.xlabel(f"распределение ср.значения оттока клиентов",fontsize = 15)

    plt.title(f'Гистограмма распределения среднего значения оттока клиентов в зависимости {parameter}', fontsize=20)

    plt.yticks(fontsize = 12)
    plt.xticks(fontsize = 12)

    return plt.show()
```

<br>**Выведем на экран гистограмму распределения факта оттока клиентов в зависимости:**

*1) от пола клиента*

```python
hist('gender', '\nот пола', 'пол клиента')
```

**Оценка с полученной гистограммы:** 
- пол клиента никак не зависит от факта его последующего оттока<br><br>

*2) возраста клиента:*

```python
hist('age', '\nот их возраста', 'возраст клиента')
```

**Оценка с полученной гистограммы:** 
- возраст клиента зависит от факта его последующего оттока, но эта зависимость незначительна<br><br>

*3) от близости проживания клиента от дома или работы:*

```python
hist('near_location', 'от\nблизости проживания от дома или работы', 'близость проживания от дома или работы')
```

**Оценка с полученной гистограммы:** 
- клиенты, которые попали в отток, чаще выбирали зал не по соседству с домом или работой по сравнению с теми, кто остался, но эта разница незначительна;
- Так как среднее значение `near_location` высока для обеих групп и разница в данном показателе для них незначительна, это говорит о развитости повсеместного расположения и доступности сети фитнес-центров «Культурист-датасаентист», потому что по собственному опыту знаю, что близость фитнесс-зала играет большую роль при его выборе

## Построим модель прогнозирования оттока клиентов

### Построим модель бинарной классификации клиентов, где целевой признак — факт оттока клиента в следующем месяце:

#### Разобьем данные на обучающую и валидационную выборку:

*Разделим наши данные на признаки (матрица X) и целевую переменную (y):*

```python
#используем датафрейм, избавленный от мультиколлениарности:
X = gym_churn_drop.drop(['churn'], axis = 1)
y = gym_churn_drop['churn']
```

*Разделим модель на обучающую и валидационную выборки:*

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
# где #random_state=0 - для сопоставления рез-тов, 
# stratify-параметр для макс. возможной установки соотношения между двумя рассматриваемыми признаками
```

*Проверим соотношение между двумя выборками:*

Соотношение между ушедшими и оставшимися равна примерно 1/3 в двух выборках. Это значит, что далее наши алгоритмы будут иметь равные возможности для изучения кажого целевого признака (факт оттока) и последующего тестирования результатов обучения

####  Оценим метрики accuracy, precision и recall для обеих моделей на валидационной выборке

*Для начала рассмотрим, что каждая метрика означает:*

- *accuracy* - доля правильных ответов

- *precision* - доля прогнозов относительно "1" класса верна

- *recall* - сколько реальных объектов "1" класса можно обнаружить с помощью модели

*Выведем на экран метрики для каждой модели:*

```python
df = pd.DataFrame ({
'accuracy': [100*accuracy_score(y_test, lr_predictions), 100*accuracy_score(y_test, rf_predictions)],
'precision': [100*precision_score(y_test, lr_predictions), 100*precision_score(y_test, rf_predictions)],
'recall': [100*recall_score(y_test, lr_predictions), 100*recall_score(y_test, rf_predictions)]})

df.index = ['LogisticRegression', 'RandomForest']

df
```
