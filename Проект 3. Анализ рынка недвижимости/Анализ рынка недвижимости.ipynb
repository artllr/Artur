```python
# Создадим таблицу с пропусками в столбце "living_area"
miss_living_area = df[df['living_area'].isna()]
print(miss_living_area['open_plan'].value_counts()) 
print(miss_living_area['studio'].value_counts()) 

print()

# Тоже самое сделаем для "kitchen_area":
miss_kitchen_area = df[df['kitchen_area'].isna()]
print(miss_kitchen_area['open_plan'].value_counts()) 
print(miss_kitchen_area['studio'].value_counts()) 
```

После анализа двух столбцов делаем вывод:

- возможно, также, что в некоторых объявлениях по продаже квартиры нет информации о жилой площади и кухни, т.к. пользователи площадки  ЯндексНедвижимость посчитали ее не столь важной, как, например, общая площадь квартиры (которая указана во всех объявлениях без исключения), либо площадь отдельных элементов квартиры указана в описании или в чертеже
<br>

Исходя из вышеперечисленных догадок, заменим пропуски в столбцах `living_area` и `kitchen_area` по среднему отношению к общей площади:
<br><br>

```python
# сооздадим два значения, описывающие соотношение жилой площади и площади кухни к общей:
ratio_living_area = df['living_area'].mean() / df['total_area'].mean()
ratio_kitchen_area = df['kitchen_area'].mean() / df['total_area'].mean()

# заполним пустые значение произведением общей площади к недавно найденным соотношением:
df['living_area'] = df['living_area'].fillna(df['total_area'] * ratio_living_area)
df['kitchen_area'] = df['kitchen_area'].fillna(df['total_area'] * ratio_kitchen_area)
```

```python
print('Медианное значение общей площади, м2:', data_centre['total_area'].median())   
print('Медианное значение стоимости квартиры, руб:', data_centre['last_price'].median())   
print('Медианное значение кол-во комнат в квартире:', data_centre['rooms'].median())  
print('Медианное значение этажа, на котором продается квартира', data_centre['floor'].median())  
print('Медианное значение этажа дома, в котором продается квартира', data_centre['floors_total'].median())   
print('Медианное значение высоты потолков', data_centre['ceiling_height'].median())  
```

**Определим корреляцию между выбранными параметрами и построим для них гистограммы:**

```python
# Определим корреляцию между параметрами:
stat_multi = data_centre.pivot_table(index='id', values=['last_price', 'total_area', 'rooms', 'ceiling_height'], aggfunc='mean')
print(stat_multi.corr())

# Определим взаимосвязь между площадью и ценой, потроив гистограмму:
data_centre.plot(x='total_area', y='last_price', kind='scatter', grid=True, alpha=0.03)

# Взаимосвязь между площадью и числом комнат:
data_centre.plot(x='total_area', y='rooms', kind='scatter', grid=True, alpha=0.3)

# Взаимосвязь между площадью и высотой потолков:
data_centre.plot(x='total_area', y='ceiling_height', kind='scatter', grid=True, alpha=0.1)

# Взаимосвязь между ценой и числом комнат:
data_centre.plot(x='last_price', y='rooms', kind='scatter', grid=True, alpha=0.02)

# Взаимосвязь между ценой и высотой потолков:
data_centre.plot(x='last_price', y='ceiling_height', kind='scatter', grid=True, alpha=0.05)

# Взаимосвязь между числом комнат и высотой потолков :
data_centre.plot(x='rooms', y='ceiling_height', kind='scatter', grid=True, alpha=0.01)
```

**Выделим также факторы, которые влияют на стоимость квартиры (число комнат, этаж, удалённость от центра):**

```python
stat_multi = data_centre.pivot_table(index='id', values=['last_price', 'rooms', 'floor', 'cityCenters_km'], aggfunc='mean')
print(stat_multi.corr()['last_price']) 

data_centre.plot(x='total_area', y='last_price', kind='scatter', legend=False, alpha=0.03, xlabel='общая площадь квартиры', 
                 ylabel='цена квартиры, 10^2 млн. руб')


stat_multi = data_centre.pivot_table(index='rooms', values='last_price', aggfunc='median')
stat_multi.plot(y='last_price', kind='bar', legend=False, xlabel='количество комнат', 
                 ylabel='цена квартиры, 10^1 млн. руб')

data_centre.plot(x='rooms', y='last_price', kind='scatter', legend=False, alpha=0.3, xlabel='этаж', 
                 ylabel='цена квартиры, 10^2 млн. руб')

floor_median = data_centre.pivot_table(index='floor_category', values='last_price', aggfunc='median')
floor_median.plot(y='last_price', kind='bar', legend=False, ylim=(0,10000000),  figsize = (3, 4), xlabel='категория этажности', 
                 ylabel='цена, млн. руб')

data_centre.plot(x='cityCenters_km', y='last_price', kind='scatter', legend=False, alpha=0.2, xlabel='расстояние до центра, км', 
                 ylabel='цена квартиры, 10^2 млн. руб')

price_median = data_centre.pivot_table(index='cityCenters_km', values='last_price', aggfunc='median')
price_median.plot(y='last_price', legend=False, xlabel='расстояние до центра, км', ylabel='цена квартиры, 10^2 млн. руб')
```

**Изучим зависимость цены квартиры в центре от даты размещения: дня недели, месяца и года:**

```python
stat_multi = data_centre.pivot_table(index='id', values=['last_price', 'datetime_week', 'datetime_month', 'datetime_year'], aggfunc='mean')
print(stat_multi.corr()['last_price'])

# Определим зависимость между ценой и датой публикации квартиры через столбчатые диаграммы:
week_median = data_centre.pivot_table(index='datetime_week', values='last_price', aggfunc='median')
week_median.plot(y='last_price', kind='bar', figsize = (4, 4), legend=False, ylim=(0, 10000000), xlabel='день недели', 
                 ylabel='цена, млн. руб')

month_median = data_centre.pivot_table(index='datetime_month', values='last_price', aggfunc='median')
month_median.plot(y='last_price', kind='bar', figsize = (4, 4), legend=False, ylim=(0, 10000000), xlabel='месяц', 
                 ylabel='цена, млн. руб')

year_median = data_centre.pivot_table(index='datetime_year', values='last_price', aggfunc='median')
year_median.plot(y='last_price', kind='bar', figsize = (4, 4), legend=False, ylim=(0, 16000000), xlabel='год', 
                 ylabel='цена, млн. руб')
```
