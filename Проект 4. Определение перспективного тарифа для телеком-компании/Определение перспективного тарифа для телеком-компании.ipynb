<br> **Далее в ходе данного этапа и проекта в целом необходимо посчитать для каждого польователя:**
- количество сделанных звонков и израсходованных минут разговора по месяцам;
- количество отправленных сообщений по месяцам;
- объем израсходованного интернет-трафика по месяцам;
- помесячную выручку с каждого пользователя (вычтите бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика; остаток умножьте на значение из тарифного плана; прибавьте абонентскую плату, соответствующую тарифному плану).


<br> **Посчитаем количество сделанных звонков и израсходованных минут разговора по месяцам для каждого пользователя:**

```python
# Создадим отдельный столбец с обозначением месяца когда происходил звонок:
calls['month'] = pd.DatetimeIndex(calls['call_date']).month

# Создадим таблицу с суммой минут разговоров по месяцам:
calls_sum = calls.pivot_table(index = ['user_id','month'], values = 'duration_min', aggfunc = 'sum')
calls_sum.columns = ['sum_duration']

# Создадим таблицу с количеством звонков по месяцам:
calls_count = calls.pivot_table(index = ['user_id','month'], values = 'duration_min', aggfunc = 'count')
calls_count.columns = ['calls_count']

display(calls_sum)
display(calls_count)
```

<br> **Посчитаем количество отправленных сообщений по месяцам для каждого пользователя:**

```python
# Создадим отдельный столбец с обозначением месяца когда было отправленно сообщение:
messages['month'] = pd.DatetimeIndex(messages['message_date']).month

# Создадим таблицу с количеством сообщений по месяцам:
messages_count = messages.pivot_table(index = ['user_id','month'], values = 'id', aggfunc = 'count')
messages_count.columns = ['messages_count']
messages_count
```

<br> **Посчитаем объем израсходованного интернет-трафика по месяцам для каждого пользователя:**

```python
internet['month'] = pd.DatetimeIndex(internet['session_date']).month
internet_count = internet.groupby(['user_id', 'month'])['mb_used'].agg(['sum'])
internet_count = internet_count.rename(columns={'sum': 'mb_used'})

#Переведем использованный трафик за месяц с Мб на Гб:
internet_count['gb_count'] = np.ceil(internet_count['mb_used'] / 1024)

display(internet_count)
```

**Объединим все полученные выше таблицы в одну:**

```python
# Используем команду join вместе с how='outer' с целью объединения всех значений, 
# даже если некоторые из них есть только в одной таблице:
total_users = calls_sum.join([calls_count, messages_count, internet_count],  how='outer').fillna(0).astype('int')
total_users = total_users.reset_index(level='month')
display(total_users.head())

# Объединим получивщуюся таблицу с таблицей "users":
df_final = users.merge(total_users, how='outer', on=['user_id'])
display(df_final.head())

# Объединим получивщуюся таблицу с таблицей "tariffs":
tariffs = tariffs.rename(columns = {"tariff_name": "tariff"})
df_final = df_final.merge(tariffs, how='outer', on='tariff')
display(df_final.head())
```

**Определим помесячную выручку с каждого пользователя:**

```python
def revenue (df_final):
    calls = df_final['sum_duration']
    messages = df_final['messages_count']
    internet = df_final['gb_count']
    tariff = df_final['tariff']
    
    if tariff == 'smart':
        i = 0
    else:
        i = 1
        
    remainder_calls = calls - tariffs['minutes_included'][i]
    if remainder_calls > 0:
        revenue_calls = remainder_calls * tariffs['rub_per_minute'][i]
    else:
        revenue_calls = 0
        
        
    remainder_messages = messages - tariffs['messages_included'][i]
    if remainder_messages > 0:
        revenue_messages = remainder_messages * tariffs['rub_per_message'][i]
    else:
        revenue_messages = 0
        
        
    remainder_internet = internet - (tariffs['mb_per_month_included'][i] / 1024)
    if remainder_internet > 0:
        revenue_internet = remainder_internet * tariffs['rub_per_gb'][i] 
    else:
        revenue_internet = 0
    
    
    revenue = revenue_calls + revenue_messages + revenue_internet + tariffs['rub_monthly_fee'][i]
    return revenue

df_final['revenue'] = df_final.apply(revenue,axis=1)
display(df_final)
```

## Шаг 4. Проверка гипотез

**<br> На данном этапе необходимо подтвердить или опровергунуть следующие гипотезы:**

- Средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
- Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

**<br> Рассмотрим первую предложенную гипотезу, которая звучит так "средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются".<br> <br>**
Для этого введем нулевую гипотезу, которая будет предполагать, что выручки с двух тарфиов равны, а далее сделаем вывод о том насколько эта гипотеза верна:

```python
from scipy import stats as st

alpha = 0.05 # определим уровень статистической значимости в пределах 5%
results = results = st.ttest_ind(
    df_smart['revenue'], df_ultra['revenue']) # выведем данную команду для проверки гипотезы о равенстве средних 
# значениях выручки с двух разных тарифов;

print('p-значение:', results.pvalue) #  выведем на экран вероятность получить наблюдаемый результат при условии, 
# что нулевая гипотеза верна

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Как и следовало догадаться: средняя выручка по двум тарифам различна. Проверим с какого тарифа выручка больше:**

```python
print(df_smart['revenue'].mean())
print(df_ultra['revenue'].mean())
print()
print(df_ultra['revenue'].mean() / df_smart['revenue'].mean())
```

*Средняя выручка пользователей тарифа «Ультра» выше чем у «Смарта» более чем на 60%*

<br>**Рассмотрим вторую предложенную гипотезу, которая звучит так "средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов":**

```python
moscow = df_final[df_final['city']=='Москва']['revenue']
 
not_moscow = df_final[df_final['city']!='Москва']['revenue']
 
# уровень значимости 5%
alpha = .05
 
results = st.ttest_ind(moscow, not_moscow, equal_var = False)
 
print('p-значение: ', results.pvalue)
 
if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<br>**Действительно: средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.** 

Проверим на сколько:

```python
print(moscow.mean())
print(not_moscow.mean())
print()
print(moscow.mean() / not_moscow.mean())
```

*Средняя выручка пользователей с Москвы выше чем с других регионов, но эта разница незначительна и из этого следует, что акцент на регионах ставить не имеет смысла*

- Первая и вторая оказались правдивыми, что подтвердили наши расчеты;
- Тариф "Ультра" оказался более прибыльным и приносит больше денег компании "Мегалайн";
